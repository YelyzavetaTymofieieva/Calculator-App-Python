```python
# stage 2 and complete code


def add(n1,n2):
    return n1+n2

def subtract(n1,n2):
    return n1-n2

def multiply(n1,n2):
    return n1*n2

def divide(n1,n2):
    return n1/n2

operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': divide,   
}

def calculate():

    continue_calculations = True
    
    try:
        num1 = float(input('Enter the first number: '))
    except ValueError:
        print('Invalid entry,please, start over.')
        return
    

    while continue_calculations:
        for symbol in operations:
            print(symbol)
        operation_symbol = input('Enter the opertion you would like to perform: ')
        
        if operation_symbol not in operations:
            print('Invalid entry,please, start over.')
            return
        try:
            num2 = float(input('Enter the second number: '))
        except ValueError:
            print('Invalid entry,please, start over.')
            return 
        
        answer = (operations[operation_symbol](num1, num2))
        print(f'{num1} {operation_symbol} {num2} = {answer}')

        should_continue = input(f"To continue calculation with {answer}, enter 'y', to restart, enter 'n'").lower()

        if should_continue == 'y':
            num1 = answer

        elif should_continue == 'n':
            continue_calculations = False
            print('\n' * 100)
            calculate()
        else:
            print('Invalid entry, start over')
            return
calculate()
```

```python
# # QUESTION  2:  Why doesn't function calculate() take any arguments? 


# # The calculate() function in your code doesn't take any arguments because it is designed to handle 
# all user input directly inside the function using input() calls. The function's role is to:

# # Prompt the user to enter numbers.
# # Prompt the user to enter an operation symbol.
# # Perform the calculation based on the user input.
# # Output the result and optionally allow the user to continue or restart.
# # Why No Arguments?
# # By not taking any arguments, the function directly interacts with the user in each step of the process. 
# This makes it easy for users to provide input interactively but does limit the flexibility of the function. 
# Here are a few reasons why this design is suitable for simple user-interactive programs:

# # User Interaction Focus: The function prompts the user for inputs every time it is called, 
# so it doesn’t need to rely on pre-existing values.
# # Self-contained Logic: All necessary inputs (numbers and operations) are gathered within the function, 
# so it does not need external data.
# # What Would Happen if It Did Take Arguments?
# # If you wanted the calculate() function to take arguments, you could design it to receive numbers 
# and the operation as parameters. This would allow you to call the function with specific values, 
# making it more versatile for use cases where you might have pre-defined numbers 
# or want to programmatically control the inputs.

```

```python
# Let's see how it would work if the claculate() took positional arguments: 

def add(n1, n2):
    return n1+n2

def subtract(n1, n2):
    return n1-n2

def multiply(n1, n2):
    return n1*n2

def divide(n1, n2):
    return n1/n2


operations = {
    '+': add,
    '-':subtract,
    '*':multiply,
    '/': divide,
}

def calculate(num1=None, operation_symbol=None, num2=None):
    # If the arguments are not provided, ask the user for input
    if num1 is None:
        try:
            num1 = float(input('Enter the first number: '))
        except ValueError:
            print('Invalid entry, please start over.')
            return

    if operation_symbol is None:
        operation_symbol = input('Enter the operation you would like to perform: ')
        if operation_symbol not in operations:
            print('Invalid operation, please start over.')
            return

    if num2 is None:
        try:
            num2 = float(input('Enter the second number: '))
        except ValueError:
            print('Invalid entry, please start over.')
            return

    # Perform the calculation
    answer = operations[operation_symbol](num1, num2)
    if answer is None:
        return

    print(f'{num1} {operation_symbol} {num2} = {answer}')

# Example of calling the function with arguments
calculate()
```

```python
#     QUESTION 3. Why do we have to call the function calculate() inside the loop and outside the function?

#                              Summary:
        
# The initial call to calculate() starts the process.
# The recursive call inside helps restart the calculator without restarting the script manually.
# Using loops instead of recursive calls can make the code cleaner and easier to manage.


# When you call calculate() outside of the function (at the end of the code), 

# this initial call starts the whole calculator process for the first time. 
# It prompts the user to enter the first number, the operation, and the second number, 
# then performs the calculation and displays the result. Without this initial call, the calculate() 
# function wouldn't run, as defining a function alone does not execute its code.


#                        Calling calculate() Inside the Function (for Restarting):
        
# Inside your loop, you have a mechanism where the user can decide whether 
# to continue using the current result or start a new calculation. 


#                         In this part:

# When the user enters 'y', the program continues the current session with the existing num1 value set 
# to the result (answer).
# When the user enters 'n', continue_calculations is set to False, ending the current loop.
# However, you then call calculate() again, which starts the function anew, 
# effectively restarting the calculator from the beginning.
# Why Call calculate() Again Inside?
# To Restart the Process: If the user chooses to start over, you need a way to reinitialize the process.
#     By calling calculate() again inside the loop when should_continue is 'n', you re-run the function, 
#     allowing the user to begin from scratch without having to manually restart the program.
# User Convenience: This makes it easy for the user to continue using the calculator indefinitely 
#     without needing to rerun the entire script. They can choose to keep calculating with the current 
#     result or start over whenever they wish.
# What Would Happen If We Didn’t Call It Inside?
# If you removed the call to calculate() inside the loop and only had the call outside, 
# the user could only perform calculations once. After they chose to stop or restart (should_continue == 'n'), 
# the program would stop without allowing them to continue with new inputs, forcing them to 
# manually restart the script.
```

```python
# example of using a while loop inside a while loop instead of usiong a def calculate function:

def add(n1, n2):
    return n1+n2

def subtract(n1, n2):
    return n1-n2

def multiply(n1, n2):
    return n1*n2

def divide(n1, n2):
    return n1/n2


operations = {
    '+': add,
    '-':subtract,
    '*':multiply,
    '/': divide,
}

continue_calc = True

num1 = float(input('1st number: '))

while continue_calc:

    for symbol in operations:
        print(symbol)
    operations_symbol = input('choose the operation: ')
    num2 = float(input('2nd number: '))
    
    result = operations[operations_symbol](num1, num2)

    print(f"{num1} {operations_symbol} {num2} = {result}")

    choice = input(f" To continue calculation with {result}, enter 'y', new  calculation, enter 'n'")

    if choice == 'y':
        num1 = result
    elif choice == 'n':
        continue_calc = False
        num1 = float(input('1st number: '))

        while continue_calc:

            for symbol in operations:
                print(symbol)
                operations_symbol = input('choose the operation: ')
                num2 = float(input('2nd number: '))

                result = operations[operations_symbol](num1, num2)

                print(f"{num1} {operations_symbol} {num2} = {result}")

                choice = input(f" To continue calculation with {result}, enter 'y', new  calculation, enter 'n'")

```
